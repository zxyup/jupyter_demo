````python
def is_prime(n):
    result = True
    for k in range(2,int(n**0.5)+1):
        if n % k == 0:
            result = False
            break
    return result

def count_primes(n)->int:
    count=0
    for k in range(2,n):
        if is_prime(k):
            count += 1
    return count
# print(is_prime(125))
print(count_primes(1000000))
````

````python
import taichi as ti
ti.init(arch=ti.cpu)

@ti.func
def is_prime(n: int):
    result = True
    for k in range(2, int(n ** 0.5) + 1):
        if n % k == 0:
            result = False
            break
    return result

@ti.kernel
def count_primes(n: int) -> int:
    count = 0
    for k in range(2, n):
        if is_prime(k):
            count += 1

    return count

print(count_primes(100000000))
````
